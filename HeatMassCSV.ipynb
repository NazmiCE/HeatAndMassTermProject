```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
import itertools
```

```python
port_size = np.linspace(0.001,0.39,100) # in m
number_of_plates = np.linspace(3
                               ,700,698)
plate_thickness = np.linspace(0.0005,0.0012,71) # in m
plate_spacing = np.linspace(0.0015,0.005,351) 
```

```python
combined_array_plate_thickness_plate_spacing = np.array(list(itertools.product(plate_thickness, plate_spacing)))
combined_array_plate_thickness_plate_spacing
```

```python
combined_array_with_port_size = np.array(list(itertools.product(combined_array_plate_thickness_plate_spacing, port_size)))
combined_array_with_port_size
```

```python
combined_array_with_port_size[:,0]
df = pd.DataFrame([combined_array_with_port_size[:,0], combined_array_with_port_size[:,1]]).T
df
```

```python
df.columns = ["fff","PORT_SIZE"]
df
```

```python
df[["THICKNESS", "PLATE_SPACING"]] = pd.DataFrame(df.fff.tolist(), index = df.index)
df.drop("fff", axis = 1, inplace = True)
```

```python
df.to_csv("DESIGN PARAMETERS COMBINED.csv")
```
